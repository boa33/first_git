```python
from sklearn.model_selection import train_test_split

X = DF1[['weight']] # 무게만 가지고 연비를 추정해보자
y = DF1[['mpg']]

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size = 0.3,
                                                    random_state = 2045)

print ('Train Data : ', X_train.shape, y_train.shape)
print ( 'Test Data : ', X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

RA = LinearRegression() # 객체를 생성
RA.fit(X_train, y_train) # 객체안의 파라미터를 트레인 데이터로 학습을 시킨다.
```

```python
print('weight(w) : ', RA.coef_)
print('bias(b) : ', RA.intercept_)
```

```python
RA.score(X_test, y_test) # 1과 가까울 수록 설명력이 좋다.
```

```python
from sklearn.metrics import mean_squared_error

y_hat = RA.predict(X_test)
mean_squared_error(y_test, y_hat) 
# MSE 값이 17이 나온다는 것은, 예측 제곱오차가 17 정도 된다는 것이다. 실제로 자동차의 무게로 연비를 추측했을 때 
# 4정도는 오차가 발생 한다는 것.
```

```python
y_hat1 = RA.predict(X)

plt.figure(figsize=(9,6))
ax1 = sns.distplot(y, hist = False, label = 'y')
ax2 = sns.distplot(y_hat1, hist = False, label = 'y_hat', ax = ax1)
plt.ylim(0, 0.07)
plt.show()

# 노랑색이 예측값, 설명된 것이 71% 정도 된다는 것
```

```python
DF2 = DF [['mpg','cylinders','horsepower','weight']]
DF2.head(3)
```

```python
from sklearn.model_selection import train_test_split

X = DF2[['weight']]
y = DF2['mpg']

X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                 test_size = 0.3, 
                                                 random_state = 2045)

print ('Trian Data : ', X_train.shape, y_train.shape)
print ('Test Data :', X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2, include_bias = False)
X_train_poly = poly.fit_transform(X_train)

print('변환 전 데이터: ', X_train.shape)
print('2차항 변환 데이터: ', X_train_poly.shape)
```

```python
from sklearn.linear_model import LinearRegression

NL = LinearRegression()
NL.fit(X_train_poly, y_train) 

```

```python
print('weight(w) :', NL.coef_)
print(' bias(b): ', '%.8f'%NL.intercept_)
```

```python
X_test_poly = poly.fit_transform(X_test)
NL.score(X_test_poly, y_test)
```

```python
from sklearn.metrics import mean_squared_error
X_test_poly = poly.fit_transform(X_test)
mean_squared_error(y_test, NL.predict(X_test_poly))
```

```python
y_hat_test = NL.predict(X_test_poly)
plt.figure(figsize=(9, 6))
plt.plot(X_train, y_train, 'o', label = 'Train Data')
plt.plot(X_test, y_hat_test, 'r+', label = 'Predicted Value')
plt.legend(loc='best')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
X_poly = poly.fit_transform(X)
y_hat2 = NL.predict(X_poly)
plt.figure(figsize = (9, 6))
ax1 = sns.distplot(y, hist=False, label="y")
ax2 = sns.distplot(y_hat2, hist=False, label="y_hat", ax=ax1)
plt.ylim(0, 0.07)
plt.show()
```
