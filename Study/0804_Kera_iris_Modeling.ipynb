```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris.info()
```

```python
iris.head()
```

```python
iris.species.value_counts()
```

```python
iris_AR = iris.values
iris_AR
```

```python
AR_X = iris_AR[:,0:4].astype(float)
AR_y = iris_AR[:,4]

AR_X.shape, AR_y.shape
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
AR_yLBE=encoder.fit_transform(AR_y)
```

```python
AR_yLBE
```

```python
from tensorflow.keras.utils import to_categorical
AR_y0HE = to_categorical(AR_yLBE)
AR_y0HE
```

```python
import tensorflow

tensorflow.__version__
```

```python
tensorflow.keras.__version__
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(AR_X, AR_y0HE,
                                                    test_size = 0.3,
                                                    random_state = 2045)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
Model_iris = models.Sequential()

Model_iris.add(layers.Dense(16, activation = 'relu', input_shape = (4,)))
Model_iris.add(layers.Dense(8, activation = 'relu'))
Model_iris.add(layers.Dense(3, activation = 'softmax')) # 다중분류이기 때문에 
# 합이 1이 되는 것
```

```python
Model_iris.summary()
```

```python
from tensorflow.keras import utils

utils.plot_model(Model_iris,
                  show_shapes = True,
                  show_dtype = True)
```

```python
Model_iris.compile(loss = 'categorical_crossentropy',
                   optimizer = 'adam',
                   metrics = ['accuracy'])
```

```python
History_iris = Model_iris.fit(X_train, y_train,
                              epochs = 500,
                              batch_size = 7,
                              validation_data = (X_test, y_test))
```

```python
import matplotlib.pyplot as plt
 

plt.figure(figsize = (9,6))
plt.ylim(0, 1.2)
plt.plot(History_iris.history['loss'])
plt.plot(History_iris.history['val_loss'])
plt.plot(History_iris.history['accuracy'])
plt.plot(History_iris.history['val_accuracy'])
plt.legend(['loss','val_loss','accuracy','val_accuracy'])
plt.grid()
plt.show()
```

```python
loss, accuracy = Model_iris.evaluate(X_test, y_test)

print ( 'Loss = {:.2f}'.format(loss))
print ( 'Accuracy = {:.2f}'.format(accuracy))
```

```python
import numpy as np
np.set_printoptions(suppress = True, precision = 5)

Model_iris.predict(X_test)
```

```python
y_hat = Model_iris.predict_classes(X_test)
y_hat
```

```python
np.argmax(Model_iris.predict(X_test), axis =1)
```

```python
y = np.argmax(y_test, axis = 1)
y
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y, y_hat)
```

```python
print(classification_report(y, y_hat,
                            target_names = ['setosa',
                                            'virginica',
                                            'versicolor']))
```
